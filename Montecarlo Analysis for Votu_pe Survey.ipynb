```python
n_preguntas = len(Question)
# Esta funcion realiza una encuesta a una persona.
def encuestar():
    matriz_personal = np.zeros(10)
    for tema in Subject:
        for pregunta in Subject[tema]:
            respuesta = random.randint(0, Question[pregunta][0]-1)
            matriz_personal += np.array(Question[pregunta][1][respuesta])
            
    return matriz_personal
```

```python
# Esta funcion realiza encuestas a un grupo de personas
# y retorna 2 arreglos,
# uno que significa las veces que el partido a salido favorito
# uno que contiene el promedio de compatibilidad del grupo encuestado.
def simulacion_parcial(n_personas: int):
    a_ini = encuestar()
    total = a_ini.reshape(1,-1)
    match = [0]*10
    match[np.argmax(a_ini)] += 1
    for i in range(n_personas-1):
        aux = encuestar()
        total = np.concatenate((total, aux.reshape(1,-1)), axis = 0)
        match[np.argmax(aux)] += 1
        
    return np.array(match), np.mean(total, axis = 0)/n_preguntas*100
        
```

```python
# Esta funcion realiza la encuesta a un grupo de personas multiples veces
# Para poder tener datos promedios y desviaciones estandar.
# La funcion entrega 4 arreglos
# El porcentaje promedio de veces que el partido sale favorito con respecto al numero de encuestas.
# La desviacion estandar del procentaje de veces que el partido sale favorito con confiabilidad 95%.
# La compatibilidad promedio con respecto al numero de encuestas.
# La desviacion estandar de la compatibilidad.

def simulacion_completa(n_encuestas, gente, conf = 1.96):
    (ganadores, compatibilidad) = simulacion_parcial(gente)
    ganadores = ganadores.reshape(1, -1)
    compatibilidad = compatibilidad.reshape(1,-1)
    for i in range(n_encuestas):
        (aux_ganador, aux_compa) = simulacion_parcial(gente)
        ganadores = np.concatenate((ganadores, aux_ganador.reshape(1, -1)))
        compatibilidad = np.concatenate((compatibilidad, aux_compa.reshape(1, -1)))
        
    a = np.mean(ganadores, axis =0)/gente*100
    b = np.std(ganadores, axis = 0)*conf/gente*100
    c = np.mean(compatibilidad, axis = 0)
    d = np.std(compatibilidad, axis = 0)*conf
    return a, b, c, d 
```

```python
%%time
# Se correla simulaci√≥n
# con n_encuestas = 10000 y gente x encuesta= 1000 demora cerca de 23 minutos 
(a,b,c,d) = simulacion_completa(100, 1000)
```
